Brand Contribution %

```python
brand_data = df['Brand'].value_counts()
plt.pie(x = brand_data,labels=brand_data.index,shadow=True,autopct='%0.2f%%')
```

The highest contribution amongst all brand is from brand Nike while lowest is Gucci in the list

```python
brand_data = df['Product_Name'].value_counts()
plt.pie(x = brand_data,labels=brand_data.index,shadow=True,autopct='%0.2f%%')
```

The highest contributor is in product category is Jeans while Sweater is at lowest

```python
a=sns.catplot(data=df,x="Category",hue="Size",kind="count",height=5,aspect=2.5)
ax = a.facet_axis(0, 0)
for c in ax.containers:
    labels = [f'{(v.get_height() / 1):.1f}' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge')
plt.title("Category vs Size")
plt.show()
```

Considering Size as a matter among kids the Small, women xl and men large tops the chart

```python
a=sns.catplot(data=df,x="Brand",hue="Color",kind="count",palette="tab20b_r",height=5,aspect=2.5)
ax = a.facet_axis(0, 0)
for c in ax.containers:
    labels = [f'{(v.get_height() / 1):.1f}' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge')
plt.title("Brand vs Color")
plt.show()
```

Considering all the brands production of articles in white remains the highest choice of color

```python
a=sns.catplot(data=df,x="Product_Name",hue="Brand",kind="count",palette="Set1",height=5,aspect=2.5)
ax = a.facet_axis(0, 0)
for c in ax.containers:
    labels = [f'{(v.get_height() / 1):.1f}' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge')
plt.title("Product vs Brand")
plt.show()
```

Considering the production of different articles Nike overall tops the choice among customers
